```python
'''
A_name:name,
A_time:time,
A_os:os,
A_osVer['FullBuildString']:osVer
A_appid:appid
A_appVer:appVer
c_data:data
'''
import json
def json_to_dict(data):
    data1 = data.encode().decode('utf-8-sig')
    json_data = json.loads(data1)
    valueList = json_data['valueList']
    ls = []
    for i in range(len(valueList)):
        dic = {}
        dic['name'] = valueList[i]['A_name']
        dic['time'] = valueList[i]['A_time']
        dic['os'] = valueList[i]['A_os']
        dic['osVer'] = valueList[i]['A_osVer']['FullBuildString']
        dic['appid'] = valueList[i]['A_appId']
        dic['appVer'] = valueList[i]['A_appVer']
        dic['data'] = valueList[i]['C_data']
        ls.append(dic)
    return ls
```
